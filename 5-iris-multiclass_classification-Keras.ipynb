```python
import pandas as pd
import numpy as np

import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dense
```

```python
data = pd.read_csv( 'C:\\Users\\Administrator\\Desktop\\Python\\AI\\Day 3\\iris.csv')
data.head(2)
```

```python
x_input  = data.drop( 'species' , axis = 1 )
y_label  = data['species']
```

```python
# Instantiating the model

mymodel = Sequential()

mymodel.add( Dense(units = 3 , input_dim = 4 , activation = 'softmax' ) )
```

```python
# compiling 

mymodel.compile( optimizer = 'adam' , loss = 'sparse_categorical_crossentropy' , metrics = ['accuracy'])

# data not in one hot form - sparse_categorical_crossentropy 
# data in one hot form - categorical_crossentropy
```

```python
mymodel.summary()
```

```python
# training

mymodel.fit( x_input , y_label , epochs = 1000 )
```

```python
mymodel.evaluate( x_input , y_label )
```
