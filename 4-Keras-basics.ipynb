```python
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dense
```

```python
data = pd.read_csv( 'data.csv' )
data.head()
```

```python
x_label = data[['x1','x2']]
y_label = data[['label']]
```

```python
x_label.head(2)
```

```python
y_label.head(2)
```

```python
plt.scatter( data.x1 , data.x2 , c = data.label)
```

```python
#Instanting the model

mymodel = Sequential()
```

```python
# mymodel.add( Dense( units = 1 , input_dim = 2 , activation = 'sigmoid' ) )

mymodel.add( Dense( units = 1 , input_dim = 2 ) )
```

```python
mymodel.summary()
```

```python
# compiling 

mymodel.compile( optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'] )
```

```python
# training 

# mymodel.fit( x_label , y_label , epochs = 1000 )

mymodel.fit( x_label , y_label , epochs = 1100)
```

```python
mymodel.evaluate( x_label , y_label )
```
