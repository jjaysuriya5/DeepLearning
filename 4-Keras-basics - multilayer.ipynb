```python
import tensorflow as tf

import pandas as pd
import numpy as np

from sklearn.metrics import accuracy_score , confusion_matrix

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dense
```

```python
data = pd.read_csv( 'data.csv' )
data.head()
```

```python
x_label = data[['x1','x2']]
y_label = data[['label']]
```

```python
x_label.head(2)
```

```python
y_label.head(2)
```

```python
plt.scatter( data.x1 , data.x2 , c = data.label)
```

```python
#Instanting the model

mymodel = Sequential()
```

```python
mymodel.add( Dense( units = 300 , input_dim = 2 , activation = 'relu' ) )
mymodel.add( Dense( units = 200  , activation = 'relu' ) )
mymodel.add( Dense( units = 50 , activation = 'relu' ) )
```

```python
mymodel.add( Dense( units = 1 , activation = 'sigmoid' ) )
```

```python
mymodel.summary()
```

```python
# compiling 

mymodel.compile( optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'] )
```

```python
# training 

mymodel.fit( x_label , y_label , epochs = 1000 )
```

```python
mymodel.evaluate( x_label , y_label )
```
